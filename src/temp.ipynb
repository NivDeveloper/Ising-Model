```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
data = []
with open("../data/20Temp.txt") as file:
    for line in file:
        data.append(line.strip("\n").split(","))


plt.figure()
plt.title("energy")
plt.scatter([float(data[x][0]) for x in range(len(data))],
            [float(data[x][1]) for x in range(len(data))],
            marker=".")

plt.figure()
plt.title("magnetization")
plt.scatter([float(data[x][0]) for x in range(len(data))],
            [float(data[x][2]) for x in range(len(data))],
            marker=".")
```

```python
data2 = []
data4 = []
data10 = []
data40 = []
meanfield = []
with open("../data/20smooth.txt") as file:
    for line in file:
        data2.append(line.strip("\n").split(","))
with open("../data/40smooth.txt") as file:
    for line in file:
        data4.append(line.strip("\n").split(","))
with open("../data/100smooth.txt") as file:
    for line in file:
        data10.append(line.strip("\n").split(","))
with open("../data/400smooth.txt") as file:
    for line in file:
        data40.append(line.strip("\n").split(","))
with open("../data/meanfield4.txt") as file:
    for line in file:
        meanfield.append(line.strip("\n").split(","))


plt.figure()
# plt.ylim(bottom=0, top=40000)
plt.title("energy")
plt.scatter([float(data2[x][0]) for x in range(len(data2))],
            [float(data2[x][1]) for x in range(len(data2))],
            marker=".")

plt.figure()
# plt.ylim(bottom=0, top = 1.05)
plt.xlim(left=0, right=5)
plt.xlabel("Temperature (K)")
plt.ylabel("Total allignment")

plt.scatter([float(data2[x][0]) for x in range(len(data2))],
            [(abs(float(data2[x][2]))/(20**2)) for x in range(len(data2))],
            c='r',
            label="20$\\times$20",
            marker=".")
plt.scatter([float(data4[x][0]) for x in range(len(data4))],
            [(abs(float(data4[x][2]))/(40**2)) for x in range(len(data4))],
            c='g',
            label="40$\\times$40",
            marker=".")
plt.scatter([float(data10[x][0]) for x in range(len(data10))],
            [(abs(float(data10[x][2]))/(100**2)) for x in range(len(data10))],
            c='b',
            label="100$\\times$100",
            marker=".")
plt.scatter([float(data40[x][0]) for x in range(len(data40))],
            [(abs(float(data40[x][2]))/(400**2)) for x in range(len(data40))],
            c='k',
            label="400$\\times$400",
            marker=".")
plt.scatter([float(meanfield[x][0]) for x in range(len(meanfield))],
            [(abs(float(meanfield[x][1]))) for x in range(len(meanfield))],
            c='m',
            label="Mean Field approx",
            marker=".")
plt.legend()

plt.axvline(2.27,
            c="c")
plt.text(1.35, 0.1,
         "T = 2.27K $\\rightarrow$")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def run(name):
    data = []
    data2 = []
    data3 = []
    data4 = []
    data5 = []
    data6 = []
    data7 = []
    with open("../data/correlations/corr" + name + "0.100000.txt") as file:
        for line in file:
            data.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "1.000000.txt") as file:
        for line in file:
            data2.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "1.500000.txt") as file:
        for line in file:
            data3.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.000000.txt") as file:
        for line in file:
            data4.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.270000.txt") as file:
        for line in file:
            data5.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.500000.txt") as file:
        for line in file:
            data6.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "3.000000.txt") as file:
        for line in file:
            data7.append(line.strip("\n").split(","))


    plt.plot([float(data[x][0]) for x in range(len(data))],
            [float(data[x][1]) for x in range(len(data))],
            c='b',
            label="T = 0.1")
#     plt.plot([float(data2[x][0]) for x in range(len(data2))],
#             [float(data2[x][1]) for x in range(len(data2))],
#             c='g',
#             label="T = 1.0")
    plt.plot([float(data3[x][0]) for x in range(len(data3))],
            [float(data3[x][1]) for x in range(len(data3))],
            c='r',
            label="T = 1.5")
    plt.plot([float(data4[x][0]) for x in range(len(data4))],
            [float(data4[x][1]) for x in range(len(data4))],
            c='m',
            label="T = 2.0")
    plt.plot([float(data5[x][0]) for x in range(len(data5))],
            [float(data5[x][1]) for x in range(len(data5))],
            c='g',
            label="T = 2.27")
    plt.plot([float(data6[x][0]) for x in range(len(data6))],
            [float(data6[x][1]) for x in range(len(data6))],
            c='c',
            label="T = 2.5")
    plt.plot([float(data7[x][0]) for x in range(len(data7))],
            [float(data7[x][1]) for x in range(len(data7))],
            c='k',
            label="T = 3.0")

    plt.legend()
    plt.ylabel("Correlation")
    plt.xlabel("Length (lattice sites)")

run("100")
```

```python
temps = [0.1, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.27, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0]
# size = 40
import math
def corrfunc(size : int, cmap : str) -> None:
    plt.figure(dpi=1500)
    corrlengths = []
    cmap = mpl.cm.get_cmap(cmap)
    for t in temps:
        data = []

        with open("../data/corrfunc/" 
                + str(size) + "/S" 
                + str(size) + "T" 
                + str(t) 
                + (("00000") if (t*10)-math.trunc(t*10) == 0 else "0000")
                + ".txt"
                ) as file:
            for line in file:
                data.append(line.strip("\n").split(","))

        line, = plt.plot([int(data[x][0]) for x in range(len(data))],
                [float(data[x][1]) for x in range(len(data))],
                label=str(t),
                c= cmap(t/4.0) if t != 2.27 else 'g'
                )
        line.set_antialiased(True)
        # corr length
        # output current temp and calculated length
        # plt.figure()
        minimum = min([float(data[x][1]) for x in range(len(data))])
        maximum = max([float(data[x][1]) for x in range(len(data))])
        thresh = minimum + (maximum-minimum)/np.e 
        for r in data:
            if float(r[1]) < thresh:
                # print(t, r[0])
                corrlengths.append([t,r[0]])
                break

    # plt.legend()
    # plt.colorbar(cmap, location='right')
    plt.xlabel("Distance $r$ (lattice sites)")
    plt.ylabel("Correlation $c(r)$")
    plt.xlim(left = 0, right = size/2 -1)
    plt.text(4.0, 0.23, "$\\uparrow T_c = 2.27\\frac{\epsilon}{k_B}$")

    plt.figure()
    plt.xlabel("Temperature $[\\frac{\epsilon}{k_B}]$")
    plt.ylabel("Correlation length (lattice sites)")
    plt.scatter([float(corrlengths[x][0]) for x in range(len(corrlengths))],
                [float(corrlengths[x][1]) for x in range(len(corrlengths))])

# corrfunc(40, "gnuplot")
```

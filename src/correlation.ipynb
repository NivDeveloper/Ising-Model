```python
import numpy as np
import matplotlib.pyplot as plt

def run(name):
    data = []
    data2 = []
    data3 = []
    data4 = []
    data5 = []
    data6 = []
    data7 = []
    with open("../data/correlations/corr" + name + "0.100000.txt") as file:
        for line in file:
            data.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "1.000000.txt") as file:
        for line in file:
            data2.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "1.500000.txt") as file:
        for line in file:
            data3.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.000000.txt") as file:
        for line in file:
            data4.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.270000.txt") as file:
        for line in file:
            data5.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "2.500000.txt") as file:
        for line in file:
            data6.append(line.strip("\n").split(","))
    with open("../data/correlations/corr" + name + "3.000000.txt") as file:
        for line in file:
            data7.append(line.strip("\n").split(","))


    plt.plot([float(data[x][0]) for x in range(len(data))],
            [float(data[x][1]) for x in range(len(data))],
            c='b',
            label="T = 0.1")
#     plt.plot([float(data2[x][0]) for x in range(len(data2))],
#             [float(data2[x][1]) for x in range(len(data2))],
#             c='g',
#             label="T = 1.0")
    plt.plot([float(data3[x][0]) for x in range(len(data3))],
            [float(data3[x][1]) for x in range(len(data3))],
            c='r',
            label="T = 1.5")
    plt.plot([float(data4[x][0]) for x in range(len(data4))],
            [float(data4[x][1]) for x in range(len(data4))],
            c='m',
            label="T = 2.0")
    plt.plot([float(data5[x][0]) for x in range(len(data5))],
            [float(data5[x][1]) for x in range(len(data5))],
            c='g',
            label="T = 2.27")
    plt.plot([float(data6[x][0]) for x in range(len(data6))],
            [float(data6[x][1]) for x in range(len(data6))],
            c='c',
            label="T = 2.5")
    plt.plot([float(data7[x][0]) for x in range(len(data7))],
            [float(data7[x][1]) for x in range(len(data7))],
            c='k',
            label="T = 3.0")

    plt.legend()
    plt.ylabel("Correlation")
    plt.xlabel("Length (lattice sites)")

run("100")
```

```python
def leng(name):
    lengths = []
    with open("../data/correlations/corrlen" + str(name) + ".txt") as file:
        for line in file:
            lengths.append(line.strip("\n").split(","))

    plt.scatter([float(lengths[x][0]) for x in range(len(lengths))],
            [float(lengths[x][1]) for x in range(len(lengths))])

leng("40")
```

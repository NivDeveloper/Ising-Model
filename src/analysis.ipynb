# spin over time with variance

```python
data2 = []
with open("../data/spinvstime.txt") as file:
    for line in file:
        data2.append(line.strip("\n").split(","))
        
#calculate variance
# 
ran = 300
var = []
for i in range(len(data2)):
    total=0
    for j in range(0 if i<ran else i-ran, i+1):
        total +=int(data2[j][1])
    mean = total/(ran)
    sumdiff = 0
    for j in range(0 if i<ran else i-ran, i+1):
        sumdiff+=(mean-int(data2[j][1]))**2
    var.append(np.sqrt(sumdiff/(ran)))

plt.figure()
plt.errorbar([int(data2[x][0]) for x in range(len(data2))],
            [(int(data2[x][1])) for x in range(len(data2))],
            yerr = var,
            ecolor="g",
            elinewidth=0.1,
            capsize=0.2)
```
